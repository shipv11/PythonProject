#### matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# bar chart showing rank wise highest salary using matplotlib

plt.bar(df["rank"], df["salary"], width= 0.5, color= "green")
plt.xlabel("Rank")
plt.ylabel("Salary")
plt.title("Rank wise salary")
plt.show
```

```python
# bar chart showing rank wise average salary using seaborn

sns.barplot(x="rank", y="salary", data=df)
plt.show
```

```python
# bar chart showing rank wise average salary using matplotlib

df.groupby("rank")[["salary"]].mean().sort_values(by="salary", ascending=False).plot(kind="bar")
plt.show
```

```python
# scatterplot showing the relationship between salary and phd for different ranks

sns.scatterplot(x="phd",y="salary",data=df,hue="rank")
plt.show
```

```python
# Pairplot showing the relationship between different numeric columns

sns.pairplot(df)
plt.show
```

```python
# Boxplot showing the distribution of salary

plt.boxplot(df[["salary"]])
plt.show
```

```python
df[["salary"]].skew()          # finding the skewness of salary
```
